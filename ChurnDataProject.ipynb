```python
# Split customers into 4 groups
male_yes = df[(df['gender']=='Male') & (df['Churn']=='Yes')]
male_no = df[(df['gender']=='Male') & (df['Churn']=='No')]
female_yes = df[(df['gender']=='Female') & (df['Churn']=='Yes')]
female_no = df[(df['gender']=='Female') & (df['Churn']=='No')]

# Print segment sizes
print(f"Male Churned: {len(male_yes)}")
print(f"Male Retained: {len(male_no)}")
print(f"Female Churned: {len(female_yes)}")
print(f"Female Retained: {len(female_no)}")
```

```python
genders = ['Male', 'Female']
no_churn = [len(male_no), len(female_no)]
churn = [len(male_yes), len(female_yes)]

# Create stacked bar chart
fig, ax = plt.subplots(figsize=(8,6))
bars_no = ax.bar(genders, no_churn, color='#2c5282', label='No Churn')
bars_yes = ax.bar(genders, churn, bottom=no_churn, color='#c53030', label='Churned')

ax.bar_label(bars_no)
ax.bar_label(bars_yes)

# Styling
ax.set_title('Customer Retention by Gender')
ax.set_ylabel('Number of Customers')
ax.legend(title='Churn Status')
plt.tight_layout()
plt.show()
```

```python
# Filter for churned customers only
churned_customers = df[df['Churn'] == 'Yes']

# Create tenure histogram
fig, ax = plt.subplots(figsize=(16, 10))
n, bins, patches = ax.hist(churned_customers['tenure'], 
                          bins=72, 
                          edgecolor='black', 
                          alpha=0.8)

for i, patch in enumerate(patches):
    if bins[i] <= 36:
        patch.set_color('red')
    else:
        patch.set_color('steelblue')

for i, count in enumerate(n):
    if count > 0:
        bin_center = (bins[i] + bins[i+1]) / 2
        ax.text(bin_center, count + 1, f'{int(count)}', 
               ha='center', va='bottom', fontweight='bold')

# Styling
ax.set_title('Customer Churn by Tenure\n(Red = High-Risk Period: 0-3 Years)')
ax.set_xlabel('Tenure (Months)')
ax.set_ylabel('Number of Churned Customers')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Print summary statistics
total_churned = len(churned_customers)
early_churned = len(churned_customers[churned_customers['tenure'] <= 36])
print(f"\nTenure Analysis Summary:")
print(f"Total churned customers: {total_churned}")
print(f"Churned in first 3 years: {early_churned} ({early_churned/total_churned*100:.1f}%)")
```
